# Using Pandas to create dummy variables

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
df_merged = pd.concat([df, dummies], axis='columns')
df_merged
```

```python
df_final = df_merged.drop(['town'], axis='columns')
df_final
```

```python
df_final = df_final.drop(['west windsor'], axis='columns')
df_final
```

```python
x = df_final.drop(['prices'], axis='columns')
x
```

```python
y = df_final.prices
y
```

```python
model = linear_model.LinearRegression()
model.fit(x, y)
```

```python
model.predict([[2600, 1, 0]])
```

# ***One hot encoding***

```python
le = LabelEncoder()
df_labelled = df
df_labelled.town = le.fit_transform(df_labelled.town)
df_labelled
```

```python
x = df_labelled[['town', 'area']].values
x
```

```python
y = df_labelled.prices
y
```

```python
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
x = ct.fit_transform(x)
x
```

```python
x = x[:, 1:]
x
```

```python
model_1 = linear_model.LinearRegression()
model_1.fit(x, y)
```

```python
model_1.predict([[0, 1, 3400]])
```

```python
model_1.score(x, y)
```
