```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../input/car-prices-2/carPrices2.csv")
df
```

```python
%matplotlib inline
plt.xlabel("Mileage")
plt.ylabel("Sell Prices")
plt.scatter(df['Mileage'], df['Sell Prices']) #Plot between Mileage and Sell Prices
```

```python
plt.xlabel("Age(in years)")
plt.ylabel("Sell Prices")
plt.scatter(df["Age_yrs"], df["Sell Prices"]) #Plot between Ages and Sell Prices
```

```python
x = df[['Age_yrs','Mileage']]
x
```

```python
y = df[['Sell Prices']]
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test, y_test)
```
