```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv("../input/salaries-below-100k/salaries_below_100k.csv")
df
```

```python
df_labelled = df
df_labelled
```

```python
le = LabelEncoder()
df_labelled.companies = le.fit_transform(df.companies)
df_labelled.job = le.fit_transform(df.job)
df_labelled.degree = le.fit_transform(df.degree)
df_labelled
```

```python
x = df_labelled.drop('salary_more_then_100k', axis = 'columns')
x
```

```python
y = df.salary_more_then_100k
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict(x_test)
```
