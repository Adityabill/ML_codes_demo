```python
df = pd.read_csv("../input/hiring-candidate-data/hiringCandidateData.csv")
df
```

```python
test_median = df.test_score.median()
test_median
```

```python
df.test_score = df.test_score.fillna(test_median)
df
```

```python
df.experience = df.experience.fillna("zero")
df
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
df.experience
```

```python
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience', 'test_score', 'interview_score']], df.salary)
```

```python
exp_in = input("Enter experience of candidate: ")
test_score_in = input("Enter test score of candidate: ")
interview_score_in = input("Enter interview score of candidate: ")
```
