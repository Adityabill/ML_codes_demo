```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../input/insurance-purchase-data/insurance_purchase.csv")
df
```

```python
%matplotlib inline
plt.xlabel("Age")
plt.ylabel("bought_insurance")
plt.scatter(df.ages, df.bought_insurance)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[['ages']], df.bought_insurance, test_size=0.3)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test, y_test)
```
